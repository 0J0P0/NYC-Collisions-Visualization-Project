```python
import os
import re
import geopy as gp
import numpy as np
import pandas as pd
import altair as alt
import geopandas as gpd

dir = './Data'
temp_pre = './Data/tmp_pre'
colission_exists = False
```

```python
weather = pd.read_csv(f'{dir}/weather_clean.csv')
```

```python
att_col = [col for col in weather.columns if re.search(r'ATT', col)]

weather.drop(columns=att_col, inplace=True)
```

## Weather Visualizations

```python
# select a random batch of the dataset (size 5000)
weather = weather.sample(5000, random_state=65)

weather.head()
```

```python
# subset of weather for 01/06/2018 - 01/07/2018
weather_2018 = weather[(weather['DATE'] >= '2018-06-01') & (weather['DATE'] <= '2018-09-31')]
weather_2020 = weather[(weather['DATE'] >= '2020-06-01') & (weather['DATE'] <= '2020-09-31')]
```

```python
# do a line plot in altair for the TOBS as a function of DATE
alt.Chart(weather_2018).mark_line(fill='lightblue').encode(
    x=alt.X('DATE:T', title='Summer 2018', axis=alt.Axis(format='%b %d', labelAngle=0)),
    y=alt.Y('TOBS:Q')
).properties(
    width=800,
    height=300
)
```
